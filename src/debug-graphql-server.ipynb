### Utils

```python
@dataclass
class CLIPFrame:
    cinemaClipEmbedding: Optional[List[float]]
    signedUrl: str

    @classmethod
    def from_graphql_node(cls, node: dict):
        data = {}

        # TODO: Get this from `self.<whatever>
        try_keys = ["cinemaClipEmbedding", "signedUrl"]

        for key in try_keys:
            data[key] = node.get(key, None)

        return cls(**data)

    def download_frame(self, save_dir: PathLike):
        download_url(self.signedUrl, str(save_dir), self.get_filename_from_url())

    def save_embedding(self, save_dir: PathLike):
        save_path = Path(save_dir) / self.get_filename_from_url()
        save_path = save_path.with_suffix(".json")
        write_json(self.cinemaClipEmbedding, save_path)

    def save(self, save_dir: PathLike):
        self.download_frame(save_dir)

        if self.cinemaClipEmbedding:
            self.save_embedding(save_dir)

    # TODO: Put this in upyog
    @staticmethod
    def _get_filename_from_url(url) -> str:
        ""
        from urllib.parse import urlparse
        url = urlparse(url)
        return Path(url.path).name

    def get_filename_from_url(self) -> str:
        return self._get_filename_from_url(self.signedUrl)
```
